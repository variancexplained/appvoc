```python
import random
from types import SimpleNamespace
from appvoc.infrastructure.web.autothrottle import AutoThrottleLatency, AutoThrottleLatencyChange
import seaborn as sns
import pandas as pd
```

## AutoThrottleLatency
This class implements a delay strategy based upon the latency.
#### Valid Response

```python

import numpy as np
lambdas = np.linspace(start=0.1,stop=1, num=10)
times = []

for l in lambdas:

    timer = AutoThrottleLatency(lambda_factor=l)
    response = {'status_code': 200}
    response = SimpleNamespace(**response)    
    delays = []


    for i in range(1000):
        latency = random.randint(200,1000) / 1000
        delay = timer.delay(latency=latency, response=response, wait=False)    
        delays.append(delay)

    ave = np.mean(delays)
    tot = sum(delays)
    d = {'lambda': l, 'Ave Delay': ave, 'Total Delay': tot}
    times.append(d)

df = pd.DataFrame(times)
df
```
