# Test EDA

```python
from appstore.infrastructure.io.local import IOService
from appstore.data.analysis.eda import EDA
```

```python
FP = "tests/data/appdata.csv"
df = IOService.read(FP)
eda = EDA(data=df)
_ = eda.histogram(x='rating', grouping='category', title="Histogram Test")
_ = eda.histogram(x="rating", title="Histogram Test")
```

## Boxplot

```python
_ = eda.boxplot(x='rating', grouping='category', title="Histogram Test")
_ = eda.boxplot(x="rating", title="Histogram Test")
```

## Barplot

```python
eda.barplot(x='category', y='rating', title='Barplot Test')
```

## Scatterplot

```python
eda.scatterplot(x='rating', y='ratings', title='Scatterplot Test')
```

## Joint Plot

```python
eda.jointplot(x='rating', y='ratings', grouping='category', title="JointPlot Test")
```

## Wordcloud

```python
eda.wordcloud(x='description', title='Description Wordcloud')
```
